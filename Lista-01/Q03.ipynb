# Exercício

Crie uma figura com TODOS os gráficos do exercício da 1ª Questão
Todos os gráficos deverão contar com Títulos, nomes dos eixos e legenda.

```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = '/content/sample_data/IRIS.csv'

data = pd.read_csv(file_path)

fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Sepallength X Petallength
axs[0, 0].plot(data['sepal_length'], data['petal_length'], color='blue', marker='o')
axs[0, 0].set_title('Sepal Length x Petal Length in Line Plot')
axs[0, 0].set_xlabel('Sepal Length')
axs[0, 0].set_ylabel('Petal Length')
axs[0, 0].legend(['Iris'])

# Sepalwidth X Petalwidth
axs[0, 1].bar(data['sepal_width'], data['petal_width'], color='green')
axs[0, 1].set_title('Sepal Width x Petal Width in Bar Plot')
axs[0, 1].set_xlabel('Sepal Width')
axs[0, 1].set_ylabel('Petal Width')
axs[0, 1].legend(['Iris'])

# Sepallength X Petalwidth
axs[1, 0].scatter(data['sepal_length'], data['petal_width'], color='red')
axs[1, 0].set_title('Sepal Length x Petal Width in Scatter Plot')
axs[1, 0].set_xlabel('Sepal Length')
axs[1, 0].set_ylabel('Petal Width')
axs[1, 0].legend(['Iris'])

# Sepalwidth X Sepallength
axs[1, 1].plot(data['sepal_width'], data['sepal_length'], color='orange')
axs[1, 1].set_title('Sepal Width x Sepal Length in Line Plot')
axs[1, 1].set_xlabel('Sepal Width')
axs[1, 1].set_ylabel('Sepal Length')
axs[1, 1].legend(['Iris'])

plt.tight_layout()
plt.show()


```
