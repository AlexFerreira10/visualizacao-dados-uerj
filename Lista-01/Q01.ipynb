# Exercício

Com o dataset `íris_csv.csv`, disponível no Teams, crie os seguintes gráficos:

1. **Sepallength X Petallength**
2. **Sepalwidth X Petalwidth**
3. **Sepallength X Petalwidth**
4. **Sepalwidth X Sepallength**

Crie os gráficos nas seguintes formas:

- Gráficos em linhas
- Gráficos de barra
- Gráficos de pontos

```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = '/content/sample_data/IRIS.csv'

data = pd.read_csv(file_path)
```

## Sepallength X Petallength

```python
# Gráfico em Linha
plt.plot(data['sepal_length'], data['petal_length'],
         color='r', marker='o')
plt.title('Sepal Length vs Petal Length in Line Plot')
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')

plt.show()
```

## Sepalwidth X Petalwidth

```python
# Gráfico em Barra
plt.bar(data['petal_width'], data['sepal_width'], color = 'r')
plt.title('Petal Length vs Sepal Width in Bar Chart')
plt.xlabel('Petal Length')
plt.ylabel('Sepal Length')

plt.show()
```

## Sepallength X Petalwidth

```python
# Gráfico em Ponto
plt.scatter(data['sepal_length'], data['petal_width'], color = 'r')
plt.title('Sepal Length vs Petal Width in Scatter Plot')
plt.xlabel('Sepal Length')
plt.ylabel('Petal Width')

plt.show()
```

## Sepalwidth X Sepallength

```python
# Gráfico em Linha
plt.plot(data['sepal_width'], data['sepal_length'], color = 'r')
plt.title('Sepal Width vs Sepal Length in Line Plot')
plt.xlabel('Sepal Width')
plt.ylabel('Sepal Length')

plt.show()
```
